```python
# reading the scraped data
df_scraped = pd.read_csv('data/htmls_scraped/scraped_data.csv')

twitter_df = pd.read_csv('data/twitter_dump_.csv')

csvs = glob.glob('candidates_csvs/*.csv')
df_candidates = pd.DataFrame()
for csv in csvs:
    # read the csv only 1,2 and 4th columns
    df_candidates = pd.concat([df_candidates, pd.read_csv(csv, usecols=[2,3,4], names=['vid_id', 'date', 'title'])])
    df_candidates = df_candidates.reset_index(drop=True)


transcripts_dict = {}
for json_file in glob.glob('data/transcripts/*.json'):
    candidate_name = json_file.split('/')[2].split('.json')[0]
    # read the json file
    with open(json_file, 'r') as f:
        json_data = json.load(f)
    for vid_id, transcript in json_data.items():
        transcripts_dict[vid_id] = transcript
df_transcripts = pd.DataFrame(list(transcripts_dict.items()), columns=['vid_id', 'text'])


# merge the transcripts with the df_candidates based on the vid_id
merged_df = pd.merge(df_candidates, df_transcripts, on='vid_id', how='left')  # or how='left', etc. as per requirement
merged_df = merged_df.dropna()

# drop vid_id column and a source column for which all the entries should be youtube
merged_df = merged_df.drop(['vid_id'], axis=1)
merged_df['source'] = 'youtube'
merged_df = merged_df.reset_index(drop=True)
merged_df
```

```python
# preprocess the text column
def preprocess_text(text):
    text = re.sub('\n+',' ',text)
    text = re.sub(' +',' ',text)
    text = re.sub('\t+',' ',text)
    text = re.sub('\r+',' ',text)
    text = text.lower()
    text = text.strip()
    return text


# Mapping of Spanish month names to English, include 3 letter words for month as well
month_mapping = {
    'enero': 'January',
    'febrero': 'February',
    'marzo': 'March',
    'abril': 'April',
    'mayo': 'May',
    'junio': 'June',
    'julio': 'July',
    'agosto': 'August',
    'septiembre': 'September',
    'octubre': 'October',
    'noviembre': 'November',
    'diciembre': 'December',
    'ene': 'January',
    'feb': 'February',
    'mar': 'March',
    'abr': 'April',
    'may': 'May',
    'jun': 'June',
    'jul': 'July',
    'ago': 'August',
    'sep': 'September',
    'oct': 'October',
    'nov': 'November',
    'dic': 'December'
}

# Mapping of Spanish day names to English in lower case letters
day_mapping = {
    'lunes': 'Monday',
    'martes': 'Tuesday',
    'mi√©rcoles': 'Wednesday',
    'jueves': 'Thursday',
    'viernes': 'Friday',
    's√°bado': 'Saturday',
    'domingo': 'Sunday'
}

# month to number mapping using three letter words
month_to_number = {
    'jan': '01',
    'feb': '02',
    'mar': '03',
    'apr': '04',
    'may': '05',
    'jun': '06',
    'jul': '07',
    'aug': '08',
    'sep': '09',
    'oct': '10',
    'nov': '11',
    'dec': '12'
}

def clean_dates(date):
    # remove newlines
    date = re.sub('\n', '', date)
    date = date.split('|')[0]
    date = date.split('T')[0]
    date = date.replace('Publicado: ', '')
    # remove a.m. and p.m.
    date = re.sub('a.m.|p.m.', '', date)
    # remove 'ES(T)'
    date = date.replace('ES', '')
    # remove the 24h time
    date = re.sub('\d{2}:\d{2}', '', date)
    # remove '-'
    date = date.replace(' - ', '')

    date = date.lower()
    # remove 'de' and commas
    date = date.replace('de', '')
    date = date.replace(',', '')
    # remove spanish days
    for day in day_mapping.keys():
        date = date.replace(day, '')
    # replace spanish months with English
    for month in month_mapping.keys():
        date = date.replace(month, month_mapping[month])
    
    date = date.lower()
    # remove extra spaces
    date = re.sub(' +', ' ', date)

    # convert 'year-month-day' pattern to this to day-month-year
    if re.search('\d{4}-\d{2}-\d{2}', date):
        date = date.split('-')
        date = date[2]+'-'+date[1]+'-'+date[0]

    date = date.strip()
    # convert day/month/year to day-month-year
    if re.search('\d{1,2}/\d{1,2}/\d{4}', date):
        date = date.split('/')
        date = date[0]+'-'+date[1]+'-'+date[2]
    
    date = date.lower()
    # # convert day Month(english) year to day-month-year
    if re.search('\d+ ([A-Za-z]+) \d+', date):
        split = date.split(' ')
        date = split[0]+'-'+month_to_number[split[1][:3]]+'-'+split[2]
    
    return date

```

```python
df['date'] = df['date'].astype(str).apply(clean_dates)
# drop all the rows with null values
df = df.dropna()
twitter_df = twitter_df.dropna()
# drop all the non-numerical values for new_dates
df = df[df['date'].progress_apply(lambda x: re.search('\d+-\d+-\d+', x)!=None)]
# convert to datetime
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
twitter_df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
```

```python
# add a new candidate column, add the name of candidates mentioned in the text
def add_candidate(text):
    candidates_list = set()
    for candidate in candidates.keys():
        for candidate_name in candidates[candidate]:
            if candidate_name in text:
                candidates_list.add(candidate)
    return candidates_list

df['candidate'] = df['text'].progress_apply(add_candidate)
```
